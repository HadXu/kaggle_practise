# 2018-04-06特征总结

```python
import pandas as pd
import numpy as np
import seaborn as sns

sns.set()
%matplotlib inline
```

```python
train = pd.read_csv('round1_ijcai_18_train_20180301.txt',delimiter=' ')
```

### 首先看下整体的转化率

```python
sns.countplot(x='is_trade',data=train)
```

```python
train.is_trade.value_counts(normalize=True)
```

> 占 0.018左右

## 考虑各个指标的转化率

### item_price_level

```python
sns.factorplot(x='is_trade',col='item_price_level',kind='count',data=train)
```

```python
percent = train.groupby(['item_price_level','is_trade']).size().transform(lambda x: x/sum(x)).reset_index().rename(columns={0:'percent'})
```

```python
percent[percent['is_trade']==1]
```

> item_price_level == 7 的时候 转化率高

### item_sales_level

```python
sns.factorplot(x='is_trade',col='item_sales_level',kind='count',data=train)
```

```python
percent = train.groupby(['item_sales_level','is_trade']).size().transform(lambda x: x/sum(x)).reset_index().rename(columns={0:'percent'})
```

```python
percent[percent['is_trade']==1]
```

> 当 item_sales_level==12 时 转化率高

```python
sns.factorplot(x='is_trade',col='item_collected_level',kind='count',data=train)
```

```python
percent = train.groupby(['item_collected_level','is_trade']).size().transform(lambda x: x/sum(x)).reset_index().rename(columns={0:'percent'})
percent[percent['is_trade']==1]
```

> 当item_collected_level == 12 13 14 比较高
